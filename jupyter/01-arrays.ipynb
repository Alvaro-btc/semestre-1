######**IMPORTANDO BIBLIOTECAS**

```python
import numpy as np # el np es un alias de la biblioteca numpy
import sys # este paquete maneja funciones de sistema
```

#### **Creando una lista de 10 elementos**

```python
lista1 = list(range(1,11))
lista1
```

#### **Consultando el tipo de dato**

```python
type(lista1)
```

#### **Transformando una lista a un Array**
Recordar que "array" es una estructura propia de la libreria numpy

```python
arreglo = np.array(lista1)
arreglo
```

#### Consultando el tipo de dato númerico del arreglo

```python
arreglo.dtype
```

#### Crear una lista de solo números unos

```python
unos = np.ones(10)
unos
```

#### **Creando lista de números ceros**

```python
ceros = np.zeros(10)
ceros
```

#### **Lista de datos compuestos (Transformando a Array)**

```python
lista2 = ['Álvaro', 2, 5.6, True]
lista2
```

#### **Transformando lista mixta en Array**

```python
arreglo2 = np.array(lista2)
arreglo2
```

#### **Consultando la cantidad de memoria ocupada por estructura (Paquete Sys)**

```python
memoria_lista = sys.getsizeof(lista1)
memoria_arreglo = arreglo.nbytes

print(f"Memoria ocupada por la lista: {memoria_lista} bytes")
print(f"Memoria ocupada por el arreglo {memoria_arreglo} bytes")
```

#### **Otros métodos de Numpy**

```python
print("Shape:",                  arreglo.shape)     # Dimensiones
print("Dimensiones:",            arreglo.ndim)      # Números de ejes
print("Tamaño:",                 arreglo.size)      # Total de elementos
print("Tipo de dato:",           arreglo.dtype)     # Tipo almacenado
print("Bytes/elem:",             arreglo.itemsize)  # Tamaño de bytes de cada elemento
```

#### **Obteniendo promedio del Arreglo**

```python
promedio = np.mean(lista1)
promedio
```

#### **Tranformando el Arreglo de Int a Float**

El método astype() transforma los elementos del arreglo a un tipo determinado de dato númerico

```python
new_arreglo = arreglo.astype(np.float32)
new_arreglo
```

```python
new_arreglo.dtype
```
