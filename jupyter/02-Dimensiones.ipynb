```python
import numpy as np
```

#### **Generando un Vector de 12 números enteros (desde el 1 al 10) con Random**

```python
vector = np.random.randint(1, 11, 12) # Tercer parametro se coloca la cantidad de elementos que tendra el vector
vector
```

#### **Obtención de la Dimensión de un Arreglo**

```python
np.shape(vector)
```

#### **Transformar el Vector de 12 elementos en una matriz 4x3**

El método reshape() redimensiona una estructura

```python
matriz = vector.reshape(4,3)
matriz
```

#### **Volviendo a consultar la diemnsión pero de la matriz**

```python
np.shape(matriz)
```

```python
import matplotlib.pyplot as plt
```

#### **Conociendo la biblioteca maplotlib**

El paquete maplotlib permite crear visualizaciones (gráficos) para interpretar resultados y datas

#### **Creando un Gráfico simple**

```python
plt.plot(vector, label = "vector")
plt.title("Grafico del Vector")
plt.xlabel("Indice (Posición)") # Agrega el titulo del eje x
plt.ylabel("Valores") # Agrega el titulo del eje y

plt.grid(True) # Para agregar una grilla al gráfico
plt.legend() # Mostrar una etiqueta, que se obtiene del label del método plot()

plt.show() # Permite visualizar gráfico
```

#### **Crear dos Vectores aleatorios en el mismo gráfico**

```python
vector1 = np.random.randint(1, 11, 12)
vector2 = np.random.randint(5, 21, 12)
```

```python
plt.plot(vector1, marker = 'o', label = "vector 1") # La función marker, agrega un marcador en los valores del vector
plt.plot(vector2, marker = 'o', label = "vector 2")

plt.title("Comparando dos Vectores")
plt.xlabel("Indice (Posición)")
plt.ylabel("Valores")

plt.grid(True)
plt.legend()

plt.show()
```

#### **Creando gráfico de barra**

```python
plt.bar(range(len(vector2)), vector2) # El método bar, perimite plotear un gráfico en barra
```
